```python
# Project2
import numpy as np
#Quadratic
def Quadratic(): 
    print("Please state the value of your (a)")
    a = float(input())
    print("Please state the value of your (b)")
    b = float(input())
    print("Please state the value of your (c)")
    c = float(input())
    
    roots = np.roots([a, b, c])
    print("The roots of your equation are:", roots)
#Cubic
def Cubic():
    print("Please state the value of your (a)")
    a = float(input())
    print("Please state the value of your (b)")
    b = float(input())
    print("Please state the value of your (c)")
    c = float(input())
    print("Please state the value of your (d)")
    d = float(input())
    
    roots = np.roots([a, b, c, d])
    print("The roots of your cubic equation are:", roots)
#Quartic
def Quartic():
    print("Please state the value of your (a)")
    a = float(input())
    print("Please state the value of your (b)")
    b = float(input())
    print("Please state the value of your (c)")
    c = float(input())
    print("Please state the value of your (d)")
    d = float(input())
    print("Please state the value of your (e)")
    e = float(input())
    
    roots = np.roots([a, b, c, d, e])
    print("The roots of your quartic equation are:", roots)

print("Please state the equation you want to find the root of: Quadratic, Cubic, or Quartic")
choice = input()
if choice == "Quadratic":
    Quadratic()
elif choice == "Cubic":
    Cubic()
elif choice == "Quartic":
    Quartic()
else:
    print("This option is not valid. Please type Quadratic, Cubic, or Quartic.")
```
